# Amazon EC2

## Build an sever and ssh to it

1. **Create a EC2 instance **
        e.g Ubuntu 64-bits, 1 GB memoery, 8 GB Disk.
2. **Add security group**
        Type  PortRange   Source
        HTTP  80          0.0.0.0/0
        SSH   22          0.0.0.0/0
        HTTPS 443         0.0.0.0/0
    * These will allow you to ssh, and use http to access to your instance
    * keep your key pair safe. ".pem" file
    
3. **SSH**
   * Open an SSH client. (find out how to connect using PuTTY)
   * Locate your private key file (ipython_notebook.pem). The wizard automatically detects the key you used to launch the instance.
   * Your key must not be publicly viewable for SSH to work. Use this command if needed:
            chmod 400 ipython_notebook.pem
   * Connect to your instance using its Public DNS:
            ec2-52-87-239-164.compute-1.amazonaws.com
              
       e.g **ssh -i "ipython_notebook.pem" ubuntu@ec2-52-87-239-164.compute-1.amazonaws.com**
